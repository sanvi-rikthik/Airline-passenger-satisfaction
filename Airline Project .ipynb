# 3. Data Cleaning

## 3.1 Dropping irrelevant columns

The dataset has a column Unnamed: 0, id which is not relevant to our analysis

```python
df=df.drop(columns=["Unnamed: 0","id"])
df
```

cleaning the cloumn headers to a standard

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('/', '_')
df.columns = df.columns.str.replace('-', '_')
```

```python
df.head(3)
```

```python
df.satisfaction.replace({'neutral or dissatisfied':0,'satisfied':1},inplace=True)
```

```python
df.satisfaction
```

## 3.2 Handling Missing Values

### Checking for the null values in each column

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
#Percentage of Null values
for col in df.columns:
    null=df[col].isnull().sum()
    percentage=(null/len(df))*100
    print(f"{col} : {percentage}%")
```

```python
# Imputing missing value with mean
df.Arrival_Delay_in_Minutes = df.Arrival_Delay_in_Minutes.fillna(df.Arrival_Delay_in_Minutes.mean())
```

```python
# Check the list of categorical variables
df.select_dtypes(include=["object"]).columns
```

```python
# Replace NaN with mode for categorical variables
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Customer_Type'] = df['Customer_Type'].fillna(df['Customer_Type'].mode()[0])
df['Type_of_Travel'] = df['Type_of_Travel'].fillna(df['Type_of_Travel'].mode()[0])
df['Class'] = df['Class'].fillna(df['Class'].mode()[0])
```

```python
df.isnull().sum()
```

## 3.3 Handling Duplicate Values

#### Checking for duplicate values

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

## 3.4 Handling Outliers

```python
df.head()
```

```python
sns.boxplot(df['Flight_Distance'])
plt.show()
```

```python
# Calculate the 25th percentile and 75th percentile
Q1=df.Flight_Distance.quantile(0.25)
Q3=df.Flight_Distance.quantile(0.75)
IQR=Q3-Q1
print(f"Q1 : {Q1}")
print(f"Q3 : {Q3}")
print(f"IQR : {IQR}")
```

```python
lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
```

```python
df=df[(df.Flight_Distance>lower_limit)&(df.Flight_Distance<upper_limit)]
```

## 12. Correlation between the Checkin service,Inflight service,On-board service , Leg room                service and satisfaction ?

```python
fig,axarr=plt.subplots(2,2,figsize=(12,8))

table1=pd.crosstab(df["satisfaction"],df["Checkin_service"])
sns.heatmap(table1,cmap='Oranges',ax=axarr[0][0])

table2=pd.crosstab(df["satisfaction"],df['Inflight_service'])
sns.heatmap(table2,cmap='Blues',ax=axarr[0][1])

table3=pd.crosstab(df["satisfaction"],df['On_board_service'])
sns.heatmap(table3,cmap='pink',ax=axarr[1][0])

table4=pd.crosstab(df["satisfaction"],df["Leg_room_service"])
sns.heatmap(table4, cmap="bone",ax=axarr[1][1])

plt.show()
```

## 13. Correlation between all the columns present in the dataset with respect to each other?

```python
# Create a correlation matrix
corr_matrix = df.corr()
corr_matrix

# Create the heatmap with annotated values in each square
plt.figure(figsize=(20, 14))  # Adjust the figure size as needed
sns.set(font_scale=1.2)  # Adjust the font size as needed
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)

# Display the plot
plt.title("Correlation Matrix")
plt.show()
```
